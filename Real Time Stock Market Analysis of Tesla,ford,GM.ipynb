```python
import yfinance as yf
import datetime

start = datetime.datetime(2020,1,1)
end = datetime.datetime(2024,12,2)

# Use yfinance to download the data
tesla = yf.download("TSLA", start=start, end=end)
ford = yf.download("F", start=start, end=end)
gm = yf.download("GM", start=start, end=end)
```

```python
tesla.head()
```

```python
tesla['Open'].plot()
```

```python
tesla['Open'].plot(label = 'Tesla',figsize=(15,7), color='blue') # Tesla in blue
gm['Open'].plot(label = 'GM', color='green') # GM in green
ford['Open'].plot(label = 'Ford', color='red') # Ford in red
plt.ylabel('Stock Price')
plt.title('Stock Price of')
plt.legend()
```

```python
tesla['Volume'].plot(label='Tesla', figsize=(15, 7), color='blue')
gm['Volume'].plot(label='GM', color='green')
ford['Volume'].plot(label='Ford', color='red')
plt.legend()
plt.title('Trading Volume Comparison')
plt.ylabel('Volume')
plt.xlabel('Date')
plt.grid(True)  # Add a grid for better readability
plt.show()
```

# Daily Percentage Change

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
```

```python
tesla.head()
```

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
gm['returns'] = (gm['Close']/gm['Close'].shift(1)) - 1
```

```python
ford.head()
```

```python
gm.head()
```

```python
ford['returns'].hist()
```

```python
tesla['returns'].hist(bins=100,label='Tesla',alpha=0)
gm['returns'].hist(bins=100,label='GM')
ford['returns'].hist(bins=100,label='Ford')
plt.legend()
```

```python
tesla['returns'].plot(kind='kde',label='Tesla')
```

```python
tesla['returns'].plot(kind='kde',label='Tesla',figsize=(15,6))
ford['returns'].plot(kind='kde',label='Ford')
gm['returns'].plot(kind='kde',label='GM')
plt.legend()
```

```python
box_df = pd.concat([tesla['returns'],gm['returns'],ford['returns']],axis = 1)
box_df.columns=['Tesla Returns','GM Returns','Ford Returns']
box_df.plot(kind='box',figsize=(16,6))
```

```python
scatter_matrix(box_df,figsize=(8,8),hist_kwds={'bins':50}, alpha = 0.1)
```

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
ford.head()
```

```python
tesla['Cumulative Return'].plot(label = 'Tesla')
```

```python
tesla['Cumulative Return'].plot(label = 'Tesla',figsize = (15,7))
ford['Cumulative Return'].plot(label = 'Ford')
gm['Cumulative Return'].plot(label = 'GM')
plt.title('Cumulative Return vs Time')
plt.legend()
```
